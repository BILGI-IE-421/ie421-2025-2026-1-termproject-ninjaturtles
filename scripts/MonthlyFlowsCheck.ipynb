```python
import pandas as pd

print("Dosya taranƒ±yor")


chunk_size = 100000  
bos_hucre_sayisi = None
toplam_satir = 0

for chunk in pd.read_csv("monthly_flows.csv", chunksize=chunk_size):
    
    if bos_hucre_sayisi is None:
        bos_hucre_sayisi = chunk.isnull().sum()
    else:
        bos_hucre_sayisi += chunk.isnull().sum()
    
    toplam_satir += len(chunk)

print(f"TARAMA TAMAMLANDI!")
print(f"Toplam Satƒ±r Sayƒ±sƒ±: {toplam_satir}")
print("S√úTUN BAZLI BO≈û H√úCRE SAYISI:")
print(bos_hucre_sayisi)


# Yorumlama
if bos_hucre_sayisi.sum() == 0:
    print("Dosyada eksik veri yok")
else:
    print("Bazƒ± h√ºcrelerde eksik veri var")
```
